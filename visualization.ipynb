```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows = 2, ncols = 2)
```

```python
fig, ax = plt.subplots(figsize = (10,4))
```

```python
data = [1,2,3,4,5,6,8,10,12]
fig, ax = plt.subplots(figsize = (6,4))
ax.set_xlabel('No of elements', fontdict = {'fontsize':15, 'color':'red'})
ax.set_ylabel('Y - Value')
ax.grid(True)
ax.set_title("Basics of Visualization", loc='right')
ax.plot(data, marker='X',markersize=10,
         markeredgecolor='yellow', markerfacecolor='cyan')
```

```python
data = [1,2,3,4,5,6,8,10,12]
data2 = [22,23,34,35,7,9,11,13]
fig,ax = plt.subplots(figsize=(5,3))
ax1.plot(data, marker='X',markersize=10,
         markeredgecolor='yellow', markerfacecolor='cyan', linestyle=':')
ax2.plot(data2, marker='D',markersize=10,
         markeredgecolor='yellow', markerfacecolor='red', linestyle='--',color='blue',linewidth=3,alpha=0.5)
```
